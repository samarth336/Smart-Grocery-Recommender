```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
# Load the dataset
df = pd.read_csv("grocery dataset.csv")

# Display the first few rows
df.head()
```

```python
# Check for missing values and data types
df.info()
df.isnull().sum()
```

```python
# Convert purchase_date to datetime
df['purchase_date'] = pd.to_datetime(df['purchase_date'])

# Extract additional time features
df['month'] = df['purchase_date'].dt.month
df['day_of_week'] = df['purchase_date'].dt.dayofweek
df['day'] = df['purchase_date'].dt.day
df['year'] = df['purchase_date'].dt.year

df.head()
```

```python
# Count number of purchases per user
user_purchase_freq = df['user_id'].value_counts().reset_index()
user_purchase_freq.columns = ['user_id', 'purchase_count']

# Merge back to the main dataframe
df = df.merge(user_purchase_freq, on='user_id', how='left')

df[['user_id', 'purchase_count']].drop_duplicates().head()
```

```python
# Check and drop duplicates
print(f"Duplicates before: {df.duplicated().sum()}")
df.drop_duplicates(inplace=True)
print(f"Duplicates after: {df.duplicated().sum()}")
```

```python
# Check unique values in season and dietary_tags
print("Seasons:", df['season'].unique())
print("\nSample dietary tags:")
print(df['dietary_tags'].sample(10).values)
```

```python
# Create a matrix where rows = users and columns = items
user_item_matrix = df.pivot_table(index='user_id', columns='item', 
                                  values='purchase_date', 
                                  aggfunc='count').fillna(0)

user_item_matrix.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Transpose the matrix to get item-item similarity
item_similarity = cosine_similarity(user_item_matrix.T)

# Create a DataFrame for easier interpretation
item_similarity_df = pd.DataFrame(item_similarity, 
                                  index=user_item_matrix.columns, 
                                  columns=user_item_matrix.columns)

item_similarity_df.head()
```

```python
def get_similar_items(item_name, top_n=5):
    if item_name not in item_similarity_df.columns:
        return f"'{item_name}' not found in dataset."
    similar_items = item_similarity_df[item_name].sort_values(ascending=False)
    return similar_items[1:top_n+1]
```

```python
def recommend_items_for_user(user_id, top_n=5):
    if user_id not in user_item_matrix.index:
        return f"User '{user_id}' not found."
    
    user_data = user_item_matrix.loc[user_id]
    purchased_items = user_data[user_data > 0].index.tolist()

    recommendations = pd.Series(dtype="float64")

    for item in purchased_items:
        similar_items = item_similarity_df[item]
        recommendations = recommendations.add(similar_items, fill_value=0)

    # Remove already purchased items
    recommendations = recommendations.drop(labels=purchased_items, errors='ignore')
    return recommendations.sort_values(ascending=False).head(top_n)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

# Convert dietary tags into list format
df['tags_list'] = df['dietary_tags'].apply(lambda x: x.split(", ") if pd.notnull(x) and x != "" else [])

# Binarize the tags
mlb = MultiLabelBinarizer()
tags_encoded = mlb.fit_transform(df['tags_list'])

# Create DataFrame of encoded tags
tags_df = pd.DataFrame(tags_encoded, columns=mlb.classes_)
df_tags = pd.concat([df[['item']], tags_df], axis=1).drop_duplicates().reset_index(drop=True)

df_tags.head()
```

```python
# Set the item as index
tag_matrix = df_tags.set_index('item')

# Compute cosine similarity between items based on dietary tags
tag_similarity = cosine_similarity(tag_matrix)

# Create similarity DataFrame
tag_similarity_df = pd.DataFrame(tag_similarity, 
                                 index=tag_matrix.index, 
                                 columns=tag_matrix.index)

tag_similarity_df.head()
```

```python
def recommend_items_by_tags(item_name, top_n=5):
    if item_name not in tag_similarity_df.columns:
        return f"'{item_name}' not found in item list."
    similar_items = tag_similarity_df[item_name].sort_values(ascending=False)
    return similar_items[1:top_n+1]
```

```python
def recommend_for_user_by_diet(user_id, top_n=5):
    if user_id not in df['user_id'].unique():
        return f"User '{user_id}' not found."
    
    # Get items purchased by the user
    user_items = df[df['user_id'] == user_id]['item'].unique()

    # Initialize recommendations
    tag_recommendations = pd.Series(dtype="float64")

    for item in user_items:
        if item in tag_similarity_df:
            tag_recommendations = tag_recommendations.add(tag_similarity_df[item], fill_value=0)

    # Remove already purchased items
    tag_recommendations = tag_recommendations.drop(labels=user_items, errors='ignore')
    return tag_recommendations.sort_values(ascending=False).head(top_n)
```

```python
# Group by season and item, and count purchases
seasonal_popularity = df.groupby(['season', 'item']).size().reset_index(name='purchase_count')

# Check top items for each season
seasonal_popularity.sort_values(['season', 'purchase_count'], ascending=[True, False]).head(10)
```

```python
def get_trending_items_by_season(season, top_n=5):
    if season not in seasonal_popularity['season'].unique():
        return f"Season '{season}' not found."
    
    trending = seasonal_popularity[seasonal_popularity['season'] == season]
    trending = trending.sort_values(by='purchase_count', ascending=False)
    return trending['item'].head(top_n).tolist()
```

```python
def recommend_seasonal_items_for_user(user_id, current_season, top_n=5):
    if user_id not in df['user_id'].unique():
        return f"User '{user_id}' not found."

    user_items = df[df['user_id'] == user_id]['item'].unique()
    
    trending_items = seasonal_popularity[seasonal_popularity['season'] == current_season]
    trending_items = trending_items.sort_values(by='purchase_count', ascending=False)

    # Exclude already purchased items
    recommendations = trending_items[~trending_items['item'].isin(user_items)]

    return recommendations['item'].head(top_n).tolist()
```

```python
def hybrid_recommendation(user_id, current_season, top_n=5, 
                          w_collab=0.4, w_content=0.4, w_season=0.2):
    if user_id not in df['user_id'].unique():
        return f"User '{user_id}' not found."

    # --- 1. Collaborative filtering scores ---
    collab_scores = pd.Series(dtype='float64')
    if user_id in user_item_matrix.index:
        user_data = user_item_matrix.loc[user_id]
        user_items = user_data[user_data > 0].index.tolist()
        for item in user_items:
            collab_scores = collab_scores.add(item_similarity_df[item], fill_value=0)
        collab_scores = collab_scores.drop(labels=user_items, errors='ignore')
    
    # --- 2. Content-based filtering scores ---
    content_scores = pd.Series(dtype='float64')
    user_items = df[df['user_id'] == user_id]['item'].unique()
    for item in user_items:
        if item in tag_similarity_df:
            content_scores = content_scores.add(tag_similarity_df[item], fill_value=0)
    content_scores = content_scores.drop(labels=user_items, errors='ignore')

    # --- 3. Seasonal popularity scores ---
    season_items = seasonal_popularity[seasonal_popularity['season'] == current_season]
    seasonal_scores = pd.Series(season_items['purchase_count'].values,
                                index=season_items['item'])
    seasonal_scores = seasonal_scores.drop(labels=user_items, errors='ignore')

    # --- Normalize all scores ---
    def normalize(s):
        return (s - s.min()) / (s.max() - s.min() + 1e-9)

    collab_scores = normalize(collab_scores)
    content_scores = normalize(content_scores)
    seasonal_scores = normalize(seasonal_scores)

    # --- Combine with weights ---
    combined_scores = (w_collab * collab_scores) \
                    .add(w_content * content_scores, fill_value=0) \
                    .add(w_season * seasonal_scores, fill_value=0)

    return combined_scores.sort_values(ascending=False).head(top_n)
```

```python
# Example: Top 5 recommendations for user U025 in Summer
hybrid_recommendation(user_id="U025", current_season="Summer", top_n=5)
```

```python
from sklearn.model_selection import train_test_split

# Create train-test split by holding out one item per user
def create_train_test_split(df):
    train_data = []
    test_data = []

    for user_id, group in df.groupby("user_id"):
        if len(group) < 2:
            train_data.append(group)
            continue
        
        test_sample = group.sample(1)
        train_sample = group.drop(test_sample.index)

        train_data.append(train_sample)
        test_data.append(test_sample)

    train_df = pd.concat(train_data)
    test_df = pd.concat(test_data)

    return train_df, test_df

train_df, test_df = create_train_test_split(df)
print("Train:", train_df.shape)
print("Test:", test_df.shape)
```

```python
# Recreate user-item matrix for training
user_item_matrix_train = train_df.pivot_table(index='user_id', columns='item', 
                                              values='purchase_date', aggfunc='count').fillna(0)
```

```python
def precision_recall_at_k(k=5):
    hits = 0
    total_actual = 0
    total_recommended = 0

    for user_id in test_df['user_id'].unique():
        actual_items = test_df[test_df['user_id'] == user_id]['item'].tolist()

        try:
            recs = hybrid_recommendation(user_id=user_id, current_season='Summer', top_n=k)
        except:
            continue

        recommended_items = recs.index.tolist()
        hit_count = len(set(actual_items) & set(recommended_items))

        hits += hit_count
        total_actual += len(actual_items)
        total_recommended += k

    precision = hits / total_recommended if total_recommended else 0
    recall = hits / total_actual if total_actual else 0
    f1 = 2 * (precision * recall) / (precision + recall + 1e-9)

    return precision, recall, f1
```

```python
precision, recall, f1 = precision_recall_at_k(k=5)
print(f"Precision@5: {precision:.4f}")
print(f"Recall@5: {recall:.4f}")
print(f"F1@5: {f1:.4f}")
```
